# Add the location columns

````python
import pandas as pd
import numpy as np
import geocoder

# Define the function to get the latitude and longtitude from postal codes
def get_lat_lng_coords(postal_code):
    lat_lng_coords = None
    
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, ontario'.format(postal_code))
        lat_lng_coords = g.latlng
    
    return lat_lng_coords[0], lat_lng_coords[1]
````

### 1. Load the ```Geospatial_Coordinates.csv```

````python
gc_df = pd.read_csv('Geospatial_Coordinates.csv')
print(gc_df.shape)
print(gc_df.head())
````

Found discrepancy in the column name __Postal Code__

````python
# Rename the first column
gc_df.rename(columns = {"Postal Code": "PostalCode"}, inplace = True)
gc_df.head()
````

### 2. Load the ```Toronto_neighborhoods.csv``` created in the previous notebook into data frame

````python
tn_df = pd.read_csv('Toronto_neighborhoods.csv')
print(tn_df.shape)
tn_df.head()
````

### 3. Merge two data frames based on ```PostalCode```

````python
total_df = pd.merge(tn_df, gc_df, on = 'PostalCode')
total_df.head()
````

### 4. Save the merged final data frame to ```Toronto_Neighborhoods_Final.csv``` 

````python
total_df.to_csv('Toronto_Neighborhoods_Final.csv', index = False)
````
